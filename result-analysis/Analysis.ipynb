```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
df = pd.read_csv("data1.csv")
```

```python
pairs = df.groupby(["subj","test_round"])["correct"].mean()
```

```python
tab = pairs.unstack()
stats.pearsonr(tab.icol(0), tab.icol(1))
```

```python
(tab.icol(0) - tab.icol(1)).mean()
```
