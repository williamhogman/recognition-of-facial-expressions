```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
df = pd.read_csv("data1.csv")
```

```python
pairs = df.groupby(["subj","test_round"])["correct"].mean()
```

```python
tab = pairs.unstack()
r, signif = stats.pearsonr(tab.icol(0), tab.icol(1))
pd.DataFrame([dict(r=r, signif=signif)])
```

```python
by_emotion = df.groupby(["subj", "emotion", "test_round", ])["correct"].mean().unstack().unstack()
```

```python
def pearson_cols(x):
    r, p = stats.pearsonr(x.icol(0), x.icol(1))
    return pd.DataFrame([dict(r=r, p=p)])

def stars(x):
    if x < 0.001:
        return "$p < 0.001$"
    elif x < 0.01:
        return "$p < 0.01$"
    elif x < 0.05:
        return "$p < 0.05$"
    elif x < 0.10:
        return "$p < 0.10$"
    else:
        return "N.S."
        

test_retest_base = by_emotion.reorder_levels([1, 0], axis=1).groupby(axis=1, level=0)
test_retest = test_retest_base.apply(pearson_cols).apply(pd.Series.round, decimals=4)
melted = pd.melt(pd.DataFrame(test_retest))
melted.columns = ("emotion", "statistic", "value")
r_and_p = melted.pivot(index="emotion", columns="statistic", values="value")[["r", "p"]]

r_and_p.index = ["Anger", "Contempt", "Disgust", "Fear", "Happiness", "Sadness", "Surprise"]
r_and_p.columns = ["Test-retest correlation ($r$)", "Significance Level"]

print r_and_p.to_latex(escape=False, index_names=False, formatters={"Significance Level": stars} )
```
